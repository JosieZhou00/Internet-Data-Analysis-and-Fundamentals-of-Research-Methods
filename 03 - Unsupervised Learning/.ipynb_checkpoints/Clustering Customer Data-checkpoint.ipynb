```python
import pandas as pd
from sklearn import preprocessing
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
customer_data = pd.read_csv("Retail Data v4 - unclean.csv")
```

```python
customer_data[0:20]
```

```python
print(customer_data.isnull().sum().sum())
```

```python
import missingno as msno
msno.matrix(customer_data)
```

```python
msno.bar(customer_data)
```

```python
#for column_name in customer_data:
#    customer_data[column_name].fillna(customer_data[column_name].mean(), inplace=True)
customer_data['Salary'].fillna(customer_data['Salary'].mean(), inplace=True)
customer_data['Annual Spend'].fillna(customer_data['Annual Spend'].mean(), inplace=True)
#print(customer_data.isnull().sum())
```

```python
print(customer_data.isnull().sum().sum())
```

```python
one_hot_gender = pd.get_dummies(customer_data['Gender'])
print(one_hot_gender[0:5])
```

```python
customer_data.drop('Gender', axis=1, inplace=True)
customer_data = one_hot_gender.join(customer_data)
customer_data[0:5]
```

```python
one_hot_married = pd.get_dummies(customer_data['Married'])
customer_data.drop('Married', axis=1, inplace=True)
customer_data = one_hot_married.join(customer_data)
```

```python
customer_data[0:5]
```

```python
from pandas.plotting import scatter_matrix
```

```python
#cd_fewer_columns = customer_data.loc[:,'Age','Salary','Annual Spend']
cd_fewer_columns = customer_data.iloc[:,4:7]
scatter_matrix(cd_fewer_columns, figsize=(16,16));
```

```python
standardized_customer_data = preprocessing.scale(customer_data)
standardized_customer_data_df = pd.DataFrame(standardized_customer_data, columns = customer_data.columns)
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(standardized_customer_data_df)
```

```python
y_km = kmeans.fit_predict(standardized_customer_data_df)
```

```python
print(y_km[0:20])
```
