```python
import csv
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
with open('Bank Reviews with cust ID v2.csv','r',encoding="ISO-8859-1") as f: 
    reader = csv.reader(f)
    complaintList = list(reader)
```

```python
df = pd.DataFrame(complaintList)
df.columns = ['customerID', 'summary','complaintText'] 
df
```

```python
vectorizer = CountVectorizer(stop_words='english', min_df =0.005) 
bagOfWords = vectorizer.fit_transform(df['complaintText']) 
vectorizer.get_feature_names()
```

```python
print(bagOfWords.toarray()[1])
```

```python
sum_words = bagOfWords.sum(axis=0)
words_freq = [(word, sum_words[0, idx]) for word, idx in vectorizer.vocabulary_.items()] 
words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
words_freq
```

```python
wordParetoDF = pd.DataFrame(words_freq, columns =['Word', 'Count']) 
wordParetoDF
```

```python
wordParetoDF[0:30].plot.bar(x='Word', y='Count')
```

```python
import nltk
from nltk.corpus import stopwords 
from nltk.corpus.reader import tagged 
from collections import OrderedDict
```

```python
stopwords_set = set(stopwords.words("english"))
```

```python
wordCount = {}
```

```python
adjectives = {'JJ', 'JJR', 'JJS'}
adverbs = {'RB', 'RBR', 'RBS'}
nouns = {'NN', 'NNS', 'NNP', 'NPS'}
verbs = {'VB', 'VBD', 'VBN', 'VBP', 'VBZ'} 
interestingTags = adjectives 
interestingTags
```

```python
tokenizer = nltk.tokenize.punkt.PunktSentenceTokenizer() 
for complaint in complaintList:
    # print('Complaint =', complaint[2])
    tokenized = tokenizer.tokenize(complaint[2]) 
    # print(tokenized)
    for toks in tokenized:
        words = nltk.word_tokenize(toks) 
        # print('Words = ', words) 
        tagged = nltk.pos_tag(words)
        # print('Tagged = ', tagged) 
        for word, gramTag in tagged:
            word = word.lower()
            # print(word, gramTag)
            if not word in stopwords_set:    # Filter out commonly occuring words
                # print("Word that is NOT a stopword = ", word)
                if gramTag in interestingTags: # Filter words only of specific types - nouns, ver 
                    wordCount.setdefault((word,gramTag),0)
                    wordCount[(word,gramTag)]+=1
```

```python
wordCount
```

```python
sortedWordCount = OrderedDict(sorted(wordCount.items(), key = lambda t: t[1], reverse=True)) 
sortedWordCount
```

```python
y = [] 
lables = []
for word, frequency in sortedWordCount.items(): 
    y.append(frequency)
    lables.append(word[0])

listStart =0
listEnd = 30
y = y[listStart : listEnd]
lables = lables[listStart : listEnd]

# print (y)
N = len(y)
x = range(N)
width = 0.7
plt.figure(figsize=(20,10))
plt.bar(x,y,width,color='cornflowerblue') # Colours at: https://matplotlib.org/examples/color/named_colors.htm 
plt.title('Adjective Frequency')
plt.xticks(x, lables, rotation=60)
plt.margins(0.01)
plt.subplots_adjust(bottom=0.2)
plt.show()
```

```python
vectorized = pd.DataFrame(bagOfWords.toarray(), columns = vectorizer.get_feature_names()) 
vectorized
```

```python
from sklearn.decomposition import NMF
```

```python
def display_topics( model, feature_names, num_top_words): 
    for ix, topic in enumerate(model.components_):
        print("\nTopic ", ix)
        print(", ".join([feature_names[i]
                         for i in topic.argsort()[:-num_top_words -1:-1]]))
```

```python
n_themes = 20         # Number of themes I want to identify
n_iterations = 1000   # Number of iterations before timing-out
n_top_words = 10      # Number of relevant words in each theme

nmf_model = NMF(n_components=n_themes, 
                random_state=1,
                alpha=.1,
                l1_ratio=.5,
                init='nndsvd',
                max_iter = n_iterations).fit(vectorized)
```

```python
display_topics(nmf_model, vectorizer.get_feature_names(), n_top_words)
```
